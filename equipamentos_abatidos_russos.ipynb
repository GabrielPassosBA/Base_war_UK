# Subistituindo o nome das colunas

```python
base_war.rename(columns={'special equipment': 'EquipamentoS'}, inplace = True)
base_war
```

```python
base_war.rename(columns={'date': 'Data'}, inplace = True)
base_war.rename(columns={'day': 'DIa'}, inplace = True)
base_war.rename(columns={'aircraft': 'Avies'}, inplace = True)
base_war.rename(columns={'helicopter': 'Helicoptero'}, inplace = True)
base_war.rename(columns={'field artillery': 'Artilharia de campo'}, inplace = True)
base_war.rename(columns={'fuel tank': 'combustivel'}, inplace = True)
base_war.rename(columns={'naval ship': 'Navios'}, inplace = True)
base_war.rename(columns={'anti-aircraft warfare': 'EqAntiAereo'}, inplace = True)
base_war
```

# Ajustando valores ausentes

```python
base_war.isnull().sum()
```

```python
media_equ_especial = base_war.EquipamentoS.mean()
base_war.fillna(media_equ_especial,inplace=True)
base_war.pop('EquipamentoS')
base_war
```

```python
base_war.pop('EquipamentoS')
base_war.pop('SBRM_sistemaMovel')
base_war
```

# Plotando visualizao de dados

```python
Plot_abates_eq_aereos = px.line(x=base_war['date'], y=base_war['aircraft'])
Plot_abates_eq_aereos.update_layout(title='Abate de Avies Russos em funo do tempo',
xaxis_title='Data',
yaxis_title='Quantidade de Avies', 
template = 'plotly_white') #adicionando o template
Plot_abates_eq_aereos
```

```python
#Importando Bibliotecas e definindo variveis
import plotly.graph_objects as go
grafico = go.Figure()
grafico.add_trace(go.Scatter(x= base_war['date'], y= base_war['aircraft'], name = 'Avies'))
grafico.add_trace(go.Scatter(x= base_war['date'], y= base_war['helicopter'], name = 'helicopteros'))
grafico.add_trace(go.Scatter(x= base_war['date'], y= base_war['drone'], name = 'Drones'))

#Adcionando templates do Grfico
grafico.update_layout(title='Abates de equipamentos areos Russos em funo do tempo',
xaxis_title='Equipamentos',
yaxis_title='Data',
plot_bgcolor = 'white',
font = {'family': 'Arial','size': 16,'color': 'black'})

#Configurando a formatao
grafico.update_xaxes( showgrid=True, gridwidth=1, gridcolor='lightgray',
showline=True, linewidth=1, linecolor='black')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray',
showline=True, linewidth=1, linecolor='black')
grafico.show()
```
